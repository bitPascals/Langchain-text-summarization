## Second summarization technique
#### Prompt Template Text Summarization

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

generic_template = """
  Write a summary of the following speech:
  Speech: {speech}
  Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template = generic_template
)

prompt
```

```python
complete_prompt = prompt.format(speech=speech, language="Spanish")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.run({"speech":speech, "language": "French"})
summary
```

## Stuff Document Chain Text Summarization

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
template = """
  Write a short and concise summary of the following speech:
  Speech: {text}"""

prompt = PromptTemplate(input_variables=["text"], template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain = load_summarize_chain(llm=llm, chain_type="stuff", prompt=prompt, verbose=True)
```

```python
output_summary = chain.run(docs)
output_summary
```

## Using Map Reduce Method to Summarixe Large Documents

```python
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100)
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
chunk_prompt = """
  Summarize the following speech below:
  Speech: {text}
  Summary:
  """

map_prompt_template = PromptTemplate(input_variables=["text"], template=chunk_prompt)
```

```python
final_prompt = """
    Provide the final and consice summary of the entire speech with the important points.
    State the summary with an introduction and conclusion.
    Speech: {text}
"""

final_prompt_template = PromptTemplate(inpput_variables=["text"], template=final_prompt)
final_prompt_template
```

```python
summary_chain = load_summarize_chain(llm=llm, chain_type="map_reduce", map_prompt=map_prompt_template, 
                                     conmbine_prompt=final_prompt_template, verbose=True)
```

```python
output = summary_chain.run(final_docs)
output
```

## Refine Chain for Summarization

```python
refine_chain_output = load_summarize_chain(llm=llm, chain_type="refine", verbose=True)
refine_chain_output.run(final_docs)
```
